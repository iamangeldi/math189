## Problem 1, Part 1

```r
baseball <- read.csv("baseball_5.csv")

# Build a linear model
model <- lm(Salary ~ Hits, data = baseball)

# PLot the data
plot(baseball$Hits, baseball$Salary,
     xlab = "Hits",
     ylab = "Salary",
     main = "Scatter plot of Hits vs Salary",
     pch = 19,
     col = "blue")

# Add the regression line
abline(model, col = "red", lwd = 2)
```

```r
# Get coefficient estimates and standard errors
coef_estimates <- summary(model)$coefficients[, "Estimate"]
standard_errors <- summary(model)$coefficients[, "Std. Error"]

# Print results
cat("Intercept Estimate:", coef_estimates["(Intercept)"], "\n")
cat("Hits Coefficient Estimate:", coef_estimates["Hits"], "\n")
cat("Standard Error for Intercept:", standard_errors["(Intercept)"], "\n")
cat("Standard Error for Hits:", standard_errors["Hits"], "\n")
```

```r
# Get RSS from model summary 
rss <- sum(residuals(model)^2)

# Get R squared
r_squared <- summary(model)$r.squared

# Print the results
cat("Residual Sum of Squares (RSS):", rss, "\n")
cat("R-squared:", r_squared, "\n")
```

Not really – the points are all over the place around that straight line. It goes up in the right direction, but there’s so much scatter that it’s a pretty weak fit.

## Problem 1, Part 2:

```r
# Build and fit a multiple linear regression model
multi_model <- lm(Salary ~ Hits + Walks + PutOuts + CHits, data = baseball)

# Get estimate coefficients and standard errors
multi_coef_estimates <- summary(multi_model)$coefficients[, "Estimate"]
multi_standard_errors <- summary(multi_model)$coefficients[, "Std. Error"]
# Print results
cat("Intercept Estimate (Multiple):", multi_coef_estimates["(Intercept)"], "\n")
cat("Hits Coefficient Estimate (Multiple):", multi_coef_estimates["Hits"], "\n")
cat("Walks Coefficient Estimate (Multiple):", multi_coef_estimates["Walks"], "\n")
cat("PutOuts Coefficient Estimate (Multiple):", multi_coef_estimates["PutOuts"], "\n")
cat("CHits Coefficient Estimate (Multiple):", multi_coef_estimates["CHits"], "\n")
cat("Standard Error for Intercept (Multiple):", multi_standard_errors["(Intercept)"], "\n")
cat("Standard Error for Hits (Multiple):", multi_standard_errors["Hits"], "\n")
cat("Standard Error for Walks (Multiple):", multi_standard_errors["Walks"], "\n")
cat("Standard Error for PutOuts (Multiple):", multi_standard_errors["PutOuts"], "\n")
cat("Standard Error for CHits (Multiple):", multi_standard_errors["CHits"], "\n")
```

```r
# Calculate RSS for multiple regression
multi_rss <- sum(residuals(multi_model)^2)
# Calculate R-squared for multiple regression
multi_r_squared <- summary(multi_model)$r.squared
# Print the results
cat("Residual Sum of Squares (RSS) (Multiple):", multi_rss, "\n")
cat("R-squared (Multiple):", multi_r_squared, "\n")
```

```r
# Extract and print p-values (marginal effects)
multi_coef <- summary(multi_model)$coefficients
p_values <- multi_coef[, "Pr(>|t|)"]
print(p_values)
```

## Problem 1, Part 3:

```r
anova(model, multi_model)
```

The ANOVA comparison yields an F-statistic of 40.7 with a p-value of approximately 1.4 × 10⁻²¹. Because this p-value is effectively zero (much smaller than 0.05), we reject the null hypothesis that “Salary depends only on Hits.” In plain terms, adding Walks, PutOuts, and CHits to the model leads to a statistically significant reduction in unexplained error. Therefore, the multivariate model is definitely a better fit for predicting Salary than the simple Hits-only model.

## Problem 2, Part b:

```r
logistic_reg <- glm(Direction ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 + Volume,
                   data = Weekly,
                   family = binomial)
summary(logistic_reg)
```

Yes, the predictor Lag2 appears to be statistically significant as the p-value for that predictor is less than the typical alpha level of 0.05.

## Problem 2, Part c:

```r
glm_probs <- predict(logistic_reg, type = "response")

# convert prob to labels
glm_pred <- ifelse(glm_probs > 0.5, "Up", "Down")

# convert labeled prob to have same levels as response
glm_pred <- factor(glm_pred, levels = levels(Weekly$Direction))

conf_matrix <- table(Predicted = glm_pred, Actual = Weekly$Direction)
conf_matrix
```

```r
# fraction of correct predictions
acc <- mean(glm_pred == Weekly$Direction)
acc
```

The confusion matrix is showing that there are 430 false negatives, 557 true positives, 54 true negatives, and 48 false positives. This suggests that the logistic regression makes a lot more type two errors where it predicts an up week while it should really be down week and that it predicts up weeks more frequently than down weeks. 

## Problem 2, Part d:

```r
# training data is from 1990 to 2008
train <- Weekly$Year <= 2008
test <- Weekly$Year > 2008
Weekly_train <- Weekly[train, ]
Weekly_test <- Weekly[test, ]

logistic_reg_2 <- glm(Direction ~ Lag2, data = Weekly_train, family = binomial)
summary(logistic_reg_2)
```

```r
probs <- predict(logistic_reg_2, Weekly_test, type = "response")

# convert prob to labels
pred <- ifelse(probs > 0.5, "Up", "Down")

# convert labeled prob to have same levels as response
pred <- factor(pred, levels = levels(Weekly$Direction))

conf_matrix <- table(Predicted = pred, Actual = Weekly_test$Direction)
conf_matrix
```

```r
# fraction of correct predictions
acc <- mean(pred == Weekly_test$Direction)
acc
```
